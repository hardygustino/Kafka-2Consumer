```python
import json, time, random
from kafka import KafkaProducer

BOOTSTRAP = "kafka:9092"
TOPIC = "events_topic"
users = ["user_1", "user_2", "user_3", "user_4", "user_5"]

producer = KafkaProducer(
    bootstrap_servers=BOOTSTRAP,
    key_serializer=lambda k: k.encode("utf-8"),
    value_serializer=lambda v: json.dumps(v).encode("utf-8"),
    acks="all"
)

print("Producer running... Ctrl+C / stop cell to stop.")
i = 0
try:
    while True:
        user = random.choice(users)        # key
        payload = {
            "user": user,
            "event": random.choice(["click", "view", "buy"]),
            "seq": i,
            "ts": time.time()
        }
        producer.send(TOPIC, key=user, value=payload)
        producer.flush()

        print(f"sent -> key={user} value={payload}")
        i += 1
        time.sleep(5)
except KeyboardInterrupt:
    print("Producer stopped.")
finally:
    producer.close()
```
